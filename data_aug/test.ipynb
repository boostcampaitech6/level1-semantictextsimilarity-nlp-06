```python
import json

with open('data.json', 'r') as f:
    dict1 = json.load(f)

with open('synonym_dict(freq>=10).json', 'r') as f:
    dict2 = json.load(f)

print(dict1)
print(dict2)
```

```python
combined_dict = {key: val+v for key, val in dict1.items() for k, v in dict2.items() if key == k}
```

```python
combined_dict
```

```python
with open('combined.json', 'w', encoding='utf-8-sig') as file:
    json.dump(combined_dict, file, ensure_ascii=False)
```
